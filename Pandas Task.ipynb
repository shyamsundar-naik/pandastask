#### 14. total percentage of active hour for a month

```python
df2 =df[['Agent','Duration']].groupby(['Agent'])['Duration'].sum()
```

```python
df3 = df2.reset_index()
df3
```

```python
d = pd.to_timedelta(df3['Duration']).dt.components[['days', 'hours', 'minutes', 'seconds']]
```

```python
df4 = pd.concat([df3,d], axis = 1)
```

```python
def convert_to_hours(delta):
    total_seconds = delta.total_seconds()
    hours = str(int(total_seconds // 3600)).zfill(2)
    return hours
```

```python
df3["Total Hours"] = df3["Duration"].apply(convert_to_hours)
```

```python
df3['Total Hours'] = df3['Total Hours'].astype('int')
```

```python
df3['Percent'] = (df3["Total Hours"] / (30 *24)) * 100
```
